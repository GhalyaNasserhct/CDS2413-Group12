```python
#Check for normality
from scipy.stats import shapiro

# select the column you want to test
risk = df['heartDiseaseRiskScore']

# Shapiro-Wilk test
stat, p = shapiro(risk)

print('stat=%.2f, p=%.30f' % (stat, p))

if p > 0.05:
    print('Normal distribution')
else:
    print('Not a normal distribution')
```

```python
# Perform Pearson's Correlation
from scipy.stats import pearsonr

r, p_value = pearsonr(df['Age'], df['heartDiseaseRiskScore'])

print(f"Pearson's Correlation Coefficient: {r:.4f}")
print(f"P_Value: {p_value:.4f}")
```

```python
from scipy.stats import spearmanr

# Spearman correlation for numerical variables
spearman_corr, spearman_p = spearmanr(df['Age'], df['heartDiseaseRiskScore'])

print(f"Spearman Correlation Coefficient: {spearman_corr:.4f}")
print(f"P-Value: {spearman_p:.4f}")

# Interpretation check
if spearman_p < 0.05:
    print("The correlation is statistically significant.")
else:
    print("The correlation is not statistically significant.")
```

```python
#chi square testing
from scipy.stats import chi2_contingency

# create a contingency table
contingency_data = pd.crosstab(df['smokingStatus'], df['riskCategory'])

# chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_data)

print('chi2=%.3f, p=%.3f' % (chi2, p))

if p > 0.05:
    print('independent categories')
else:
    print('dependent categories')
```

```python
#One Sample T-test
from scipy.stats import ttest_1samp

# Dependent variable
risk = df['heartDiseaseRiskScore']


pop_mean = 60

t_stat, p_value = ttest_1samp(risk, pop_mean)

print("One-Sample T-Test Results")
print(f"Sample Mean = {risk.mean():.4f}")
print(f"Population Mean (Hypothesized) = {pop_mean}")
print(f"T-statistic = {t_stat:.4f}")
print(f"P-value = {p_value:.4f}")

if p_value < 0.05:
    print("Reject H0: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject H0: No significant difference between sample mean and population mean.")
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
print(y_kmeans)
```

```python
plt.scatter(X.iloc[y_kmeans == 0, 0], X.iloc[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X.iloc[y_kmeans == 1, 0], X.iloc[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X.iloc[y_kmeans == 2, 0], X.iloc[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 250, c = 'yellow', label = 'Centroids')
plt.title('K-Means Clustering: Age vs Cholesterol Level')
plt.xlabel('Age')
plt.ylabel('Cholesterol (mg/dL)')
plt.legend()
plt.show()
```
