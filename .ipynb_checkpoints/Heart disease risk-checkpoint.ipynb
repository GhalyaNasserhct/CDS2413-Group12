```python
#Check for normality
from scipy.stats import shapiro

# select the column you want to test
risk = df['heart_disease_risk_score']

# Shapiro-Wilk test
stat, p = shapiro(risk)

print('stat=%.2f, p=%.30f' % (stat, p))

if p > 0.05:
    print('Normal distribution')
else:
    print('Not a normal distribution')
```

```python
# Perform Pearson's Correlation
from scipy.stats import pearsonr

r, p_value = pearsonr(df['age'], df['heart_disease_risk_score'])

print(f"Pearson's Correlation Coefficient: {r:.4f}")
print(f"P_Value: {p_value:.4f}")
```

```python
#chi square testing
from scipy.stats import chi2_contingency

# create a contingency table
contingency_data = pd.crosstab(df['smoking_status'], df['risk_category'])

# chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_data)

print('chi2=%.3f, p=%.3f' % (chi2, p))

if p > 0.05:
    print('independent categories')
else:
    print('dependent categories')
```
